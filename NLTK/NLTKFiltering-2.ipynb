```python
import pandas as pd
data = pd.read_csv("data1.csv")
data = data.drop(['header_text', 'contentLength', 'body_Length', 'header_length', 'Title'], axis=1)

import re
def fix_new_line(txt):
    newS = re.sub(r"\n"," ", txt)
    return newS

data["body_Text"] = (data["body_Text"].dropna().apply(lambda x: fix_new_line(x)))


punct = '!"#$%&\()*+,-.:;<=>?@[\\]^_`{|}~'
def remove_punctuation(txt):
    new_text = "".join([c for c in txt if c not in punct])
    return new_text

data["body_Text"] = (data["body_Text"].dropna().apply(lambda x: remove_punctuation(x)))

def remove_countries(txt):
    new_text = "".join([c for c in txt if not re.search('/[a-zA-Z]{2}/',txt)])
    return new_text

data["originalUrl"] = (data["originalUrl"].dropna().apply(lambda x: remove_countries(x)))

import nltk
import string
from nltk.tokenize import TweetTokenizer
tknzr = TweetTokenizer()
ascii_chars = set(string.printable) 
def remove_non_ascii(txt):
    new_text = " ".join([w for w in tknzr.tokenize(txt) if all(char in ascii_chars for char in w)])
    return new_text

data["body_Text"] = (data["body_Text"].dropna().apply(lambda x: remove_non_ascii(x)))


# filters
filter = data["originalUrl"] != ""
data = data[filter]
filter = data["body_Text"] != "NaN"
data = data[filter]
filter = data["body_Text"] != ""
data = data[filter]


def fix_front_slash(txt):
    newS = re.sub(r"\/"," ", txt)
    return newS

data["body_Text"] = (data["body_Text"].dropna().apply(lambda x: fix_front_slash(x)))


def fix_quotes(txt):
    newS = re.sub('"','', txt)
    return newS
data["body_Text"] = (data["body_Text"].dropna().apply(lambda x: fix_quotes(x)))

data.dropna(subset=['body_Text'], inplace=True)
```
